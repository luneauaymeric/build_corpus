```python
df = pd.read_json('data.raw.json')
```

```python
d = df.loc[df.label.str.contains('Ukr')]
emo = d.emoji.iloc[0]
u = html.unescape(emo)
u.encode('utf-8')
```

```python
dict_html= {}

for n, h in enumerate(df.hexcode):
    emoji = df.emoji.iloc[n].encode('utf-8')
    dict_html[h] = emoji
dict_html

df["html_code"] = df.hexcode.map(dict_html.get)
```

```python
df = df[['label', 'hexcode', 'emoji', 'html_code']]
df.to_json('emoji_html.json', index = False)
```
